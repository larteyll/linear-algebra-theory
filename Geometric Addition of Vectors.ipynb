## plot the addition using geometric method 

```python
import numpy as np 
import matplotlib.pyplot as plt 

# Create vectors
np.random.seed(42)
v1 = np.array([3,-8])
v2 = np.linspace(6,12,2)
v3 = v1 + v2

# Plot v1 from origin
plt.plot([0, v1[0]], [0, v1[1]], 'b', label="v1")

# Plot v2 from the end of v1
plt.plot([v1[0], v1[0] + v2[0]], [v1[1], v1[1] + v2[1]], 'r', label="v2")

# Plot v3 from origin
plt.plot([0, v3[0]], [0, v3[1]], 'k', label="v3 (v1+v2)")

# Indicate x and y axes
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

# Formatting
plt.legend()
plt.grid(True)

# Change the length of the x and y axes
plt.xlim(-20, 20)  # Set limits for the x-axis
plt.ylim(-20, 20)  # Set limits for the y-axis

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Geometric Addition of Vectors')
plt.show()
```
